Neste exemplo:

Estamos utilizando o modelo BLIP (Bootstrapping Language-Image Pre-training) da Salesforce, que é um modelo multimodal para geração de legendas de imagens.
Carregamos o processador e o modelo usando a biblioteca transformers.
Carregamos uma imagem de exemplo e a preparamos para o modelo.
Geramos uma legenda para a imagem e a imprimimos.

```python
from transformers import BlipProcessor, BlipForConditionalGeneration
from PIL import Image

# Carregar o processador e o modelo BLIP (que é um modelo multimodal)
processor = BlipProcessor.from_pretrained("Salesforce/blip-image-captioning-base")
model = BlipForConditionalGeneration.from_pretrained("Salesforce/blip-image-captioning-base")

# Carregar uma imagem de exemplo
image_path = "coca.jpg"  # Substitua pelo caminho da sua imagem
image = Image.open(image_path).convert("RGB")

# Preparar a imagem para o modelo
inputs = processor(image, return_tensors="pt")

# Gerar a legenda
```

```python
outputs = model.generate(**inputs)
caption = processor.decode(outputs[0], skip_special_tokens=True)

print("Legenda gerada:", caption)
```
