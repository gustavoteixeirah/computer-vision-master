```python
import time
```

```python
# Multiplicao matricial com listas
lista = [1]*10_000_000

start = time.process_time()

dot = 0
for elemento in lista:
  dot += elemento * elemento

stop = time.process_time()

print("Resultado da operao matricial:", dot)
tempo_decorrido = stop - start
print('O tempo decorrido foi de', tempo_decorrido, "segundos")
```

```python
# Multiplicao matricial com array numpy
vetor = np.asarray(lista) #  OU a = np.ones(100000)

start = time.process_time()

dot = vetor.dot(vetor)

stop = time.process_time()

print("Resultado da operao matricial:", dot)
tempo_decorrido_2 = stop - start
print('O tempo decorrido foi de', tempo_decorrido_2, "segundos")
```
