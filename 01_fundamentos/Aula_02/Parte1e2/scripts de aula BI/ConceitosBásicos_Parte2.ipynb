# Estruturas condicionais

```python
import numpy as np
num = np.random.normal()
num
```

```python
if num > 1:
  print('Número é maior que 1')
```

```python
if num > 1:
  print('Número é maior que 1')
else:
  print("Número menor ou igual a 1")
```

```python
if num > 1:
  print('Número é maior que 1')
elif num > -1:
  print('Número está entre -1 e 1')
else:
  print('Número menor que -1')
```

```python
# ifs aninhados
num = 2.1
if num > 1:
  print("Maior que 1")
  if num > 2:
    print("Maior que 2")
```

#Estruturas de repetição

## WHILE

```python
while False:
  print('Hello')
```

```python
while True:
  print('Hello')

# cuidado com loops infinitos!
```

```python
contador = 1
while contador < 5:
  print('Hello')
  contador += 1  # contador = contador + 1
```

##FOR

```python
for i in [3, 45, 32]:
  print(i)
```

```python
for i in range(50):
  print(i, end=' ')
```

```python
# for aninhado
for i in range(3):
  for j in range(2):
    print(i, j)
```

```python
vetor = np.random.normal(size=3)
vetor
```

```python
# for aninhado
for i in vetor:
  for j in range(2):
    print(i, j)
```
